```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"C:\Users\vorap\OneDrive\Desktop\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
sneakers=driver.find_element(By.CLASS_NAME,"_3704LK")
sneakers.send_keys("sneakers")
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
brand=[]
price=[]
product=[]

brands=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in brands:
    b=i.text
    brand.append(b)
    
prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
for i in prices:
    p=i.text
    price.append(p)
    
products=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
for i in products:
    des=i.text
    product.append(des)
```

```python
print(len(brand),len(price),len(product))
```

```python
df=pd.DataFrame({'Brand':brand,'Price':price,'Product_Description':product})
df
```
