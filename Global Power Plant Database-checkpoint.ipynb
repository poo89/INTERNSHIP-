# cross validation

```python
from sklearn.model_selection import cross_val_score

np.random.seed(10)
def rmse_cv(model, x,y):
    rmse =- (cross_val_score(model, x,y, scoring='neg_mean_squared_error', cv=10))
    return(rmse)


models = [LinearRegression(),
             Ridge(),
             SVR(kernel='linear'),
             SVR(kernel='poly'),
             SVR(kernel='rbf'),
             RandomForestRegressor(),
             DecisionTreeRegressor(),  
             GradientBoostingRegressor(),]
            


names = ['LR','R','svr','svr_p','svr_r','RF','DTR','GBR']

for model,name in zip(models,names):
    score = rmse_cv(model,x,y)
    print("{}    : {:.6f}, {:4f}".format(name,score.mean(),score.std()))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Create a RandomForestRegressor
RF = RandomForestRegressor()

# Define the parameter grid
param = {
    'n_estimators': [100, 200],
    'criterion': ['friedman_mse', 'mae'],  # Use 'friedman_mse' instead of 'mse'
    # Add other hyperparameters you want to tune
}

# Create the GridSearchCV object
RF_grid = GridSearchCV(estimator=RF, param_grid=param, cv=10, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)

# Fit the grid search to your data
RF_grid.fit(x_train, y_train)

# Get the best estimator and make predictions
best_RF = RF_grid.best_estimator_
RF_grid_pred = best_RF.predict(x_test)
```
