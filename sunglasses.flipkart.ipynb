```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://www.flipkart.com/")
```

```python
sunglasses=driver.find_element(By.CLASS_NAME,"_3704LK")
sunglasses.send_keys("sunglasses")
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
brand=[]
description=[]
price=[]
start=0
end=3
for page in range(start,end):

    brandsun=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in brandsun:
    brands=i.text
    brand.append(brands)    
    descrip=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
for i in descrip:
    des=i.text
    description.append(des)

    pri = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
for i in pri:
    p=i.text
    price.append(p)
    
    

                                  
```

```python
next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
next_button.click()
```

```python
print(len(brand),len(description),len(price))
```

```python
df=pd.DataFrame({"Brand":brand,"Product_Description":description,"Price":price})
df
```
